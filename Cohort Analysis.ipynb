```python
import numpy as np
import pandas as pd
from datetime import datetime

file_path = r'C:\Users\smorr\Desktop\CV portfolio etc\2023 - New practical projects\Python\Cohort_Analysis\KPMG_VI_New_raw_data_update_final.xlsx'
df1 = pd.read_excel(file_path, sheet_name='Transactions')

```

```python
df1
```

```python
# removing the first row of the data frame to replace the header and permanently drop original header
df1.columns = df1.iloc[0]
df1.drop(df1.index[0],inplace=True, axis = 0)

# only want the Approved orders 
df1 = df1[df1['order_status'] == 'Approved']
df1
```

```python
# creating a new dataframe based on the selected columns below from df1
df2 = df1[['customer_id','transaction_date']].copy(deep=True)
df2
```

```python
# creating a new column with just year and month from the transaction date
df2['YM']= df2.apply(lambda x: int(datetime.strftime(x.transaction_date, "%Y%m")), axis=1)
df2
```

```python
# calculating the first entry in the year/month column
start_month = min(df2.YM)


df2['transaction_month_index'] = df2.YM.apply(lambda x: x - start_month)
df2
```

```python
# finding the earliest transaction month for each customer
df3 = df2.groupby('customer_id')['transaction_month_index'].min().reset_index()

# renaming column
df3 = df3.rename(columns={'transaction_month_index':'cohort_month'})

# each customer's first transaction is assigned to a cohort number (0 represents January 2017 - upwards in 1 month increments)
df3
```

```python
# changing the index to Customer Id and joining df2 and df2 together into a new df
df4 = df2.join(df3.set_index('customer_id'), on='customer_id')
df4
```

```python
"""
Creating a new column from taking the 'transaction_month_index' values 
(which represent the number of months since the start_month for each transaction) 
and subtracting the 'cohort_month' values (which represent the earliest transaction month for each customer) 

This calculation indicates how many months after the cohort month each transaction occurred for each customer.
"""

df4['cohort_index']=df4.transaction_month_index - df4.cohort_month
df4
```

```python
# keeping only the needed columns 
df5 = df4[['customer_id','cohort_month','cohort_index']]
df5
```

```python
# removing multiple transactions by each customer in the subsequent months from the cohort month (keep first and drop others)
df6 = df5.drop_duplicates(subset=['customer_id', 'cohort_month','cohort_index'], keep='first')
df6
```

```python
# counting total distinct customers for the cohort month and the difference from the cohort month (cohort_index) 
df7 = df6.groupby(['cohort_month','cohort_index'])['customer_id'].count().reset_index()
df7 = df7.rename(columns={'customer_id':'customer_count'})
df7
```

```python
"""" 
Analysing customer retention - 

reading vertically says how many customers from the original cohort (0/0) are still engaging. Following down to create
seperate cohorts to see the difference between each

reading horizontally says how customers are still active for each month on from the original cohort month
"""

# creating a Cohort Chart 
df8 = df7.pivot_table(index = 'cohort_month', columns = 'cohort_index',values = 'customer_count')
df8
```

```python
# cohort index 2 / cohort month 9 - indicates a data quality issue as the number goes up from the previous month 
# creating a copy of df8
# replacing incorrect value with the value from previous cohort index 
df9 = df8.copy()

df9.iloc[9, 2] = 5

df9
```

```python
# visualing the above Cohort Chart with a heatmap 

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1,1,figsize = (16,10))

sns.heatmap(df8, annot = True, cmap="YlGnBu", fmt='g')

ax.xaxis.tick_top() 
ax.set_xlabel('Cohort Index', fontsize=14)
ax.set_ylabel('Cohort Month', fontsize=14)
ax.xaxis.set_label_position('top')

plt.show()
```
